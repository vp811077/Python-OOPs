18. What is multiple inheritance in Python?

Ans: **Multiple Inheritance in Python** is when a **class can inherit attributes and methods from more than one parent class**.

### Key Points:

1. A child class can access **methods and variables from multiple parent classes**.
2. Python resolves conflicts using the **Method Resolution Order (MRO)**, which determines the order in which parent classes are searched when calling a method.
3. It allows **code reuse** from multiple classes but should be used carefully to avoid complexity.

### Syntax Example:

```python
class Father:
    def skills(self):
        print("Gardening")

class Mother:
    def skills(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skills()  # Output: Gardening
```

âœ… **Explanation:**

* `Child` inherits from both `Father` and `Mother`.
* Since both parents have a `skills` method, Python uses **MRO** and calls `Father`â€™s method first.

**Quick Tip:**

* Multiple inheritance = child gets features from **more than one parent**.
* Always check MRO if methods overlap.

```python
        self.age = 0

        self.age = age

    def greet(self):
        print(f"Hi, I'm {self.name} and I'm {self.age} years old")
        return self  # returns the object

```

**Output:**

```

```

---

### **Explanation:**

* `.greet()` is then called on the same object in the same line.
* This allows **cleaner, more fluent code**.

âœ… **Memory Tip:**

* Always **return `self`** in the methods to enable chaining.

````python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
# from it and implement the area() method in both
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius


# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# Object creation
c = Circle(5)
r = Rectangle(4, 6)

print("Area of Circle:", c.area())
print("Area of Rectangle:", r.area())
````

````python
#4. Demonstrate polymorphism by creating a base class Bird with a method fly().Create two derived classes Sparrow and Penguin that override the fly() method.
# Base class
class Bird:
    def fly(self):
        print("Bird can fly")


# Derived class Sparrow
class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high in the sky")


# Derived class Penguin
class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly, but it swims")


# Function to demonstrate polymorphism
def show_fly(bird):
    bird.fly()


# Object creation
b = Bird()
s = Sparrow()
p = Penguin()

show_fly(b)
show_fly(s)
show_fly(p)
````

````python
#5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
class BankAccount:
    def __init__(self, balance):
        # Private attribute
        self.__balance = balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Amount deposited:", amount)
        else:
            print("Invalid deposit amount")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print("Amount withdrawn:", amount)
        else:
            print("Insufficient balance or invalid amount")

    # Method to check balance
    def check_balance(self):
        return self.__balance


# Object creation
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)

print("Current Balance:", account.check_balance())

````

````python
#6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().
# Base class
class Instrument:
    def play(self):
        print("Instrument is being played")


# Derived class Guitar
class Guitar(Instrument):
    def play(self):
        print("Guitar is playing melodious strings")


# Derived class Piano
class Piano(Instrument):
    def play(self):
        print("Piano is playing soothing music")


# Runtime polymorphism demonstration
instruments = [Guitar(), Piano()]

for instrument in instruments:
    instrument.play()

````

````python
#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.
class MathOperations:

    # Class method
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    # Static method
    @staticmethod
    def subtract_numbers(a, b):
        return a - b


# Method calls
sum_result = MathOperations.add_numbers(10, 5)
sub_result = MathOperations.subtract_numbers(10, 5)

print("Addition:", sum_result)
print("Subtraction:", sub_result)
````

````python
#9. Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator".
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    # Overriding __str__ method
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"


# Object creation
f1 = Fraction(3, 4)
f2 = Fraction(5, 8)

print(f1)
print(f2)
````

````python
#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"


# Object creation
v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Vector Sum:", v3)
````

````python
#14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.
class Employee:
    def calculate_salary(self, hours_worked, hourly_rate):
        return hours_worked * hourly_rate


class Manager(Employee):
    def calculate_salary(self, hours_worked, hourly_rate, bonus):
        basic_salary = super().calculate_salary(hours_worked, hourly_rate)
        return basic_salary + bonus


# Example usage
emp = Employee()
print("Employee Salary:", emp.calculate_salary(40, 500))

mgr = Manager()
print("Manager Salary:", mgr.calculate_salary(40, 500, 10000))
````

````python
#16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass


class Cow(Animal):
    def sound(self):
        print("Cow says: Moo")


class Sheep(Animal):
    def sound(self):
        print("Sheep says: Baa")


# Example usage
cow = Cow()
cow.sound()

sheep = Sheep()
sheep.sound()
````

````python
#18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.
# Base class
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def display_info(self):
        print(f"Address: {self.address}")
        print(f"Price: ${self.price}")

# Derived class
class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)  # Initialize base class attributes
        self.number_of_rooms = number_of_rooms

    def display_info(self):
        super().display_info()  # Call base class method
        print(f"Number of Rooms: {self.number_of_rooms}")

# Example usage
house1 = House("123 Maple Street", 250000)
house1.display_info()

print("\n")

mansion1 = Mansion("456 Oak Avenue", 1500000, 10)
mansion1.display_info()
````
